```python
import aster.monitoring
import datetime
import pandas
import matplotlib.pyplot as plt
import seaborn
seaborn.set_style("whitegrid")
seaborn.despine()
%matplotlib inline
%config InlineBackend.figure_formats = {'svg',}
```

```python
logparser = aster.monitoring.LogParser('results/logs.csv')
```

```python
logparser.list_keys()
```

```python
# Parse logs
timeserie = logparser.get(['get_booster_earth', 'put_booster_earth'])

# Plot timeserie
plt.figure(figsize=(11, 4))
plt.plot(timeserie)
logparser.plot_launching_window()
plt.ylabel('Earth boosters')
plt.xlabel('Time')
plt.show()
```
